```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Dummy', 'KNN-1', 'KNN-2', 'KNN-3-1', 'KNN-3-2']
stages = ['Dummy Model', 'KNN Training', 'KNN Prediction', 'KNN Retraining', 'KNN Retrained Prediction']
accuracy = [0.5075, 0.895, 0.963, 0.721, 0.965]
precision = [0.5048, 0.8867, 0.9407, 0.4163, 0.913]
recall = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]

fig, ax = plt.subplots(figsize=(12, 7))
x = np.arange(len(models))
ax.plot(x, accuracy, 'o-', color='blue', linewidth=2, label='Accuracy')
ax.plot(x, precision, 's-', color='green', linewidth=2, label='Precision')
ax.plot(x, recall, '^-', color='red', linewidth=2, label='Recall')

for i in range(1, len(x)):
    ax.axvline(x=i-0.5, color='gray', linestyle='--', alpha=0.5)

ax.set_title('Model Performance Across Different Stages', fontsize=16)
ax.set_xlabel('Model Stages', fontsize=14)
ax.set_ylabel('Performance Metric', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(stages, rotation=15, ha='right')
ax.set_ylim(0, 1.05)
ax.grid(True, linestyle='--', alpha=0.7)

for i, stage in enumerate(stages):
    ax.text(i, 0.05, f"Stage {i+1}", ha='center', fontweight='bold')
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
